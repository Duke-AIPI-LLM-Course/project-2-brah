```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
import os

```

```python

all_urls = duke_websites + aipi_accessible_websites + pratt_accessible_websites
chunks = duke_descriptions + aipi_descriptions + pratt_descriptions
```

```python

def store_data(chunks):

    # 加载预训练的句向量模型（MiniLM 示例）
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')


    # 计算每个文本块的向量表示
    embeddings = model.encode(chunks)  # shape: (num_chunks, 384)
    embeddings = np.array(embeddings, dtype="float32")

    # 创建索引对象，构建 FAISS 索引（使用余弦相似度，将向量先归一化然后用内积检索）
    vec_dim = embeddings.shape[1]  # 向量维度，例如 all-MiniLM-L6-v2 为384
    index = faiss.IndexFlatIP(vec_dim)
    # 若使用余弦相似，需要先对向量进行L2归一化
    faiss.normalize_L2(embeddings)
    index.add(embeddings)
    # Store the index to disk
    faiss.write_index(index, "my_faiss.index")
    
    return index
```

```python
import os
import re
from urllib.parse import urlparse

def download_html_files(urls_list):
    """
    从URL列表中下载所有网站的HTML内容并保存为文件
    文件名基于网站域名和路径生成
    """
    # 创建保存HTML文件的目录
    save_dir = './html_files'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    for url in urls_list:
        try:
            # 获取网站HTML内容
            response = requests.get(url, timeout=15)
            if response.status_code == 200:
                # 从URL生成文件名
                parsed_url = urlparse(url)
                # 使用域名和路径创建文件名，去除http://和www.
                domain = parsed_url.netloc.replace('www.', '')
                # 处理路径部分，替换/为_
                path = parsed_url.path.strip('/')
                if path:
                    filename = f"{domain}_{path.replace('/', '_')}.html"
                else:
                    filename = f"{domain}.html"
                
                # 确保文件名合法（移除不允许的字符）
                filename = re.sub(r'[\\/*?:"<>|]', "_", filename)
                
                # 保存HTML内容到文件
                file_path = os.path.join(save_dir, filename)
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(response.text)
                print(f"已保存: {filename}")
            else:
                print(f"无法下载 {url}，状态码: {response.status_code}")
        except Exception as e:
            print(f"下载错误 {url}: {str(e)}")
    
    print(f"所有HTML文件已保存到 {save_dir}")
```

```python
import requests
from bs4 import BeautifulSoup
import os
import re
from urllib.parse import urlparse
```

```python

def extract_domain(url):
    """从URL中提取域名作为文件名"""
    parsed_url = urlparse(url)
    domain = parsed_url.netloc
    path = parsed_url.path.strip('/')
    
    if path:
        # 移除特殊字符，只保留字母、数字、下划线和连字符
        safe_path = re.sub(r'[^\w\-]', '_', path)
        return f"{domain}_{safe_path}"
    return domain

def extract_text_from_url(url):
    """从URL中提取所有文本内容"""
    try:
        # 发送HTTP请求
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()  # 如果请求不成功则抛出异常
        
        # 使用BeautifulSoup解析HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 删除script和style元素
        for script_or_style in soup(['script', 'style', 'meta', 'noscript']):
            script_or_style.decompose()
            
        # 获取文本
        text = soup.get_text(separator='\n')
        
        # 清理文本(删除多余空行和空格)
        lines = (line.strip() for line in text.splitlines())
        chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
        text = '\n'.join(chunk for chunk in chunks if chunk)
        
        return text
    except Exception as e:
        print(f"提取文本时出错 ({url}): {e}")
        return f"Error extracting content: {str(e)}"

def save_website_texts(urls):
    """从URL列表中提取文本并保存为txt文件"""
    # 创建保存文件的目录
    output_dir = "website_texts"
    os.makedirs(output_dir, exist_ok=True)
    
    for i, url in enumerate(urls):
        print(f"正在处理 [{i+1}/{len(urls)}]: {url}")
        # 获取文本内容
        text_content = extract_text_from_url(url)
        
        # 创建文件名
        filename = extract_domain(url) + ".txt"
        file_path = os.path.join(output_dir, filename)
        
        # 保存文本到文件
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(text_content)
            
        print(f"✓ 已保存到 {file_path}")

# 使用代码中的URL列表
# 假设all_urls已经定义好了
save_website_texts(all_urls)
```

```python

def save_lists_to_files(urls, descriptions, urls_file_path, descriptions_file_path):
  """保存URL和描述列表到文件
  
  Args:
      urls: URL列表
      descriptions: 描述列表
      urls_file_path: 保存URL的文件路径
      descriptions_file_path: 保存描述的文件路径
  """
  # 保存URL列表
  with open(urls_file_path, 'w', encoding='utf-8') as f:
    for url in urls:
      f.write(f"{url}\n")
  
  # 保存描述列表
  with open(descriptions_file_path, 'w', encoding='utf-8') as f:
    for desc in descriptions:
      f.write(f"{desc}\n---ITEM_SEPARATOR---\n")  # 使用分隔符，因为描述可能包含多行
```
